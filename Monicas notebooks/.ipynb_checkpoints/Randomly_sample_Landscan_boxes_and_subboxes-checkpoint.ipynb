### Calculate state level population statistics - Used to check whether there would be a variance issues with sampling PPS

```python
#Calculate some state level statistics to check variability in population distributions across Landscan boxes in each state

states = ['bihar', 'UP', 'rajasthan', 'MP', 'odisha', 'jharkand', 'assam']
stats_df = pd.DataFrame(columns = ['state','number_boxes', 'total_pop', 'mean_box_pop', 'std_dev_box_pop'])
for counter, state in enumerate(states):
    gdf = gpd.read_file("Output/state_landscan_output/{}_landscan.shp".format(state))
    display(gdf.head())
    if gdf.shape[1] == 3:
        gdf.columns = ['population', 'state', 'geometry']
    else:
        gdf.columns = ['population', 'geometry']
    
    row = pd.DataFrame({'state':state, 'number_boxes': [gdf.shape[0]], 'total_pop': [gdf.population.sum()], 'mean_box_pop': [gdf.population.mean()], 'std_dev_box_pop':[gdf.population.std()]})
    stats_df = stats_df.append(row, sort = False)
    
    #PLot histogram
    gdf.population.hist(bins = 50)
    plt.title('{} Population Histogram'.format(state.upper()))
    plt.show()
    #plt.savefig('{}_hist.png'.format(state))
    
    #Export csv of data
    #gdf['population'].to_csv('{}_landscan.csv'.format(state))
    
stats_df.head()

stats_df.to_csv('Output/state_level_landscan_stats.csv')
```

### Concatenate all state Landscan shapefiles together into one dataframe to be used for the rest of the sampling

```python
#Concatenate all state Landscan population estimates 
states = ['bihar', 'UP', 'rajasthan', 'MP', 'odisha', 'jharkand', 'assam']
all_dfs = []
for counter, state in enumerate(states):
    gdf = gpd.read_file("Output/state_landscan_output/{}_landscan.shp".format(state))
    if gdf.shape[1] == 3:
        gdf.columns = ['population', 'state', 'geometry']
        gdf = gdf.drop(['state'], axis = 1)
    else:
        gdf.columns = ['population', 'geometry']
        
    print(gdf.shape)
        
    gdf['state'] = state
    
    all_dfs.append(gdf)
    
all_landscan_NITI = pd.concat(all_dfs, axis = 0, sort = False)
all_landscan_NITI.reset_index(inplace = True, drop = True)
all_landscan_NITI['unique_id'] = all_landscan_NITI.index.values
display(all_landscan_NITI.head())
```

```python
#Some checks to tell whether it worked (should return an empty dataframe)
all_landscan_NITI['test'] = all_landscan_NITI.duplicated(['unique_id'], keep = False)
display(all_landscan_NITI[all_landscan_NITI.test])
print(all_landscan_NITI.shape)
```

```python
#Change all jharkand mispellings
all_landscan_NITI.replace('jharkand', 'jharkhand', inplace = True)

#Export dataset - This is a csv containing a list of all 1,355,188 Landscan boxes in all 7 NITI states (the full sampling frame)
all_landscan_NITI[['unique_id', 'population', 'state']].to_csv('Output/all_NITI_landscan.csv')
```

### Automatically sampling subboxes

```python
#Check that box areas are all consistently about 1km^2
def calc_area(geom):
    geom_area = ops.transform(partial(pyproj.transform,pyproj.Proj(init='EPSG:4326'),pyproj.Proj(proj='aea',lat1=geom.bounds[1],lat2=geom.bounds[3])),geom)
    return(geom_area.area)
plot_gdf['area_km'] = plot_gdf.apply(lambda x: calc_area(x.geometry)/1000000, axis = 1)

plot_gdf.area_km.hist()
```

```python
#Extract coordinates on boundary of all Landscan boxes
plot_gdf['boundary'] = plot_gdf.apply(lambda x: x.geometry.boundary.coords, axis = 1)
```

```python
#Check that for every geometry the last point is the same as the first point and that they all form squares as expected
#True except for 3 boxes near district boundaries that are odd shapes (these all have num_subboxes == 1 though so not a problem)

column = plot_gdf.boundary
for i in range(len(column)):
    print(i)
    zipped_list = list(zip(column[i].xy[1],column[i].xy[0]))
    print(zipped_list)
    lats_diff = zipped_list[0][0] - zipped_list[4][0]
    longs_diff = zipped_list[0][1] - zipped_list[4][1]

    if (lats_diff == 0) & (longs_diff == 0):
        print('')
    else:
        print("NOT SAME")
    
    plt.scatter(*zip(*zipped_list))
    plt.show()
```

```python
#Extract all boxes with num_subboxes != 1
to_segment_gdf = plot_gdf[plot_gdf.num_subboxes != 1].reset_index()
print(to_segment_gdf.shape)

#Extract all boxes with num_subboxes == 1
not_segment_gdf = plot_gdf[plot_gdf.num_subboxes == 1].reset_index()
```

```python
#Throw out last coordinate in boundary list for each polygon (since it's redundant) and reverse order of tuples since QGIS puts them in opposite order
to_segment_gdf['corners'] = to_segment_gdf.apply(lambda x: [(tuple_item[1], tuple_item[0]) for tuple_item in list(x.boundary)[:-1]], axis = 1)
to_segment_gdf['corner_lats'] = to_segment_gdf.apply(lambda x: [tuple_item[0] for tuple_item in x.corners], axis = 1)
to_segment_gdf['corner_longs'] = to_segment_gdf.apply(lambda x: [tuple_item[1] for tuple_item in x.corners], axis = 1)

```

```python
#Calculate top left x and top left y coordinate of each Landscan box as well as the len in the x and y direction
to_segment_gdf['n'] = to_segment_gdf.apply(lambda x: int(math.sqrt(x.num_subboxes)), axis = 1)

to_segment_gdf['tl_x'] = to_segment_gdf.apply(lambda x: min(x.corner_lats), axis = 1)
to_segment_gdf['tl_y'] = to_segment_gdf.apply(lambda x: max(x.corner_longs), axis = 1)
to_segment_gdf['tr_x'] = to_segment_gdf.apply(lambda x: max(x.corner_lats), axis = 1)
to_segment_gdf['bl_y'] = to_segment_gdf.apply(lambda x: min(x.corner_longs), axis = 1)

to_segment_gdf['x_len'] = to_segment_gdf.apply(lambda x: x.tr_x - x.tl_x, axis = 1)
to_segment_gdf['y_len'] = to_segment_gdf.apply(lambda x: x.tl_y - x.bl_y, axis = 1)
```

```python
#Calculate length and width of each subbox
to_segment_gdf['sbox_x_len'] = to_segment_gdf.apply(lambda x: x.x_len / x.n, axis = 1)
to_segment_gdf['sbox_y_len'] = to_segment_gdf.apply(lambda x: x.y_len / x.n, axis = 1)
```

```python
#Just to check that segmentation is working correctly
to_segment_gdf['x_calc'] = to_segment_gdf.apply(lambda x: ((x.subbox - 1) % x.n), axis = 1)
to_segment_gdf['y_calc'] = to_segment_gdf.apply(lambda x: (math.floor((x.subbox - 1)/x.n)), axis = 1)

#Calculate x and y coordinates of the corners of the selected box
to_segment_gdf['sbox_tl_x'] = to_segment_gdf.apply(lambda x: x.tl_x + x.sbox_x_len * ((x.subbox - 1) % x.n), axis = 1)
to_segment_gdf['sbox_tl_y'] = to_segment_gdf.apply(lambda x: x.tl_y - x.sbox_y_len * (math.floor((x.subbox - 1)/x.n)), axis = 1)
to_segment_gdf['sbox_tr_x'] = to_segment_gdf.apply(lambda x: x.sbox_tl_x + x.sbox_x_len, axis = 1)
to_segment_gdf['sbox_bl_y'] = to_segment_gdf.apply(lambda x: x.sbox_tl_y - x.sbox_y_len, axis = 1)
```

```python
#Put together x and y coordinates into corner tuples
to_segment_gdf['new_corners'] = to_segment_gdf.apply(lambda x: [(x.sbox_tl_x, x.sbox_tl_y), (x.sbox_tr_x, x.sbox_tl_y),(x.sbox_tr_x, x.sbox_bl_y), (x.sbox_tl_x, x.sbox_bl_y) ], axis = 1)
to_segment_gdf['new_box'] = to_segment_gdf.apply(lambda x: Polygon([(tuple_item[1], tuple_item[0]) for tuple_item in x.new_corners]), axis = 1)

to_segment_gdf['new_and_old_corners'] = to_segment_gdf['corners'] + to_segment_gdf['new_corners']
```
