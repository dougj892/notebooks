```python
import pandas as pd
import numpy as np
from tabula import read_pdf
```

```python
path = "http://img.asercentre.org/docs/Publications/ASER%20Reports/ASER%20TOT/State%20pages%20English/"
states = ["andhrapradesh", "arunachalpradesh","bihar","chhattisgarh", "gujarat", "haryana", "himachalpradesh", 
          "jammuandkashmir", "jharkhand", "karnataka", "madhyapradesh", "maharashtra", "manipur", "meghalaya", 
          "mizoram", "nagaland", "odisha", "punjab", "rajasthan", "tamilnadu", "tripura", 
          "uttarpradesh", "uttarakhand", "westbengal"]


df = pd.DataFrame(columns=['year','std3_reading1','std3_subtraction','state'])
years = np.arange(2006,2015)
```

```python
for state in states:
    print(state)
    file = path+state+".pdf"
    # for now, i am just importing table 4 and table 8. 
    all_tables = read_pdf(file, pages =[2,3], multiple_tables  = True )
    print(all_tables)
```

```python
for state in states:
    print(state)
    file = path+state+".pdf"
    # for now, i am just importing table 4 and table 8. 
    all_tables = read_pdf(file, pages =[2,3], multiple_tables  = True )
    # extract table 4 state data. i should probably break up this code, but even broken up it would still be unreadable
    # best strategy is just to perform manual checks
    table4_state_data = all_tables[0].iloc[5:14,0].to_frame()[0].str.split(expand = True)[2]
    # extract table 8 data.  since this table starts from 2007, add a NA for 2006
    table8_state_data = all_tables[2].iloc[5:14,0].to_frame()[0].str.split(expand = True)[2]
    table8_list = table8_state_data.tolist()
    table8_list.insert(0,'NA')
    temp_df = pd.DataFrame({'year':years,'std3_reading1':table4_state_data, 'std3_subtraction':table8_list,'state':[states[0]]*len(years)})
    df = df.append(temp_df)
```
