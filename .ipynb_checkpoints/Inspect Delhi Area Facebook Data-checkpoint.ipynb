```python
import pandas as pd
import numpy as np
import geopandas as gpd
import shapely
import os
import folium
%matplotlib inline
```

```python
os.chdir(r"C:\Users\dougj\Documents\GIS")
df = pd.read_csv("as43_delhi_area.csv")
```

```python
# the coordinates below are for a point in Humayunpur 
humayunpur = [28.560514, 77.195045]

# generate dataframe of points in the Humayunpur area
width = .005
h_points = df[(df.longitude > (humayunpur[1]-width)) & (df.longitude < (humayunpur[1]+width)) & 
              (df.latitude > (humayunpur[0]-width))& (df.latitude < (humayunpur[0]+width))]
```

```python
# generate a ap showing the humayunpur points
map = folium.Map(location = humayunpur, zoom_start = 12)
feature_group = folium.FeatureGroup("Locations")

for index, row in h_points.iterrows():
    feature_group.add_child(folium.Marker(location=[row.latitude, row.longitude],popup=row.population_2020))
map.add_child(feature_group)
```

```python
# generate clusters of micro-segments by dividing up into 5x5 boxes
lat_min = df.latitude.min()
long_min = df.longitude.min()

# pixel width in degrees from Facebook data 
pixel_width = 0.0002777777777779999848
span = pixel_width*5
df['lat_pixel']=(df.latitude-lat_min)/pixel_width
df['long_pixel']=(df.longitude-lat_min)/pixel_width
df['lat_group']=((df.latitude-lat_min)/span).round(decimals=0).astype(int)
df['long_group']=((df.longitude-long_min)/span).round(decimals=0).astype(int)
```

```python
# count the number of micro-segments per cluster and population per cluster
cluster_totals = df.groupby(['lat_group','long_group']).agg({'latitude':'count', 'population_2020':'sum'})
cluster_totals.rename(columns ={'latitude': 'num_segments', 'population_2020': 'total_pop'}, inplace = True)

# check that there are no clusters with more than 25 micro-segments
cluster_totals.num_segments.max()
```

```python
cluster_totals['log_pop']=np.log(cluster_totals.total_pop)
```

```python
# pick and plot a random cluster
import random
rnd_lat_group = random.randint(0, df.lat_group.max())
rnd_long_group = random.randint(0, df.long_group.max())
print("Lat group = ", rnd_lat_group)
print("Long group =",rnd_long_group)

# create a map of all the cluster points
cluster_points = df[(df.lat_group == rnd_lat_group) & (df.long_group == rnd_long_group)]
print("Number of segments in cluster =", len(cluster_points))


loc = [cluster_points.iloc[0].latitude, cluster_points.iloc[0].longitude]
map = folium.Map(location = loc, zoom_start = 12)
for index, row in cluster_points.iterrows():
    feature_group.add_child(folium.Marker(location=[row.latitude, row.longitude],popup=row.population_2020))
map.add_child(feature_group)
```
