Plot the first 20 points from the geotiff file using folium

```python
import folium
map = folium.Map(location = [kenya.Lat[0], kenya.Lon[0]], zoom_start = 13)

feature_group = folium.FeatureGroup("Locations")

for index, row in kenya[:20].iterrows():
    feature_group.add_child(folium.Marker(location=[row.Lat,row.Lon],popup=row.Population))
map.add_child(feature_group)
```
